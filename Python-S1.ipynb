## 1 — Boolean Functions

A **boolean expression** is any code that evaluates to `True` or `False`. Python provides several built-in functions that return a boolean value.

---

### `bool()` — Convert any value to True / False

| Value passed | Returns | Why |
|---|---|---|
| Any non-zero number | `True` | Has quantity |
| Any non-empty string | `True` | Has content |
| `0` | `False` | No quantity |
| `""` (empty string) | `False` | No content |
| `None` | `False` | No value at all |

```python
bool(42)        # True
bool("hello")   # True
bool(0)         # False
bool("")        # False
bool(None)      # False
```

> **Use case:** Quickly check whether a variable holds any meaningful data before processing it.

---

### `all()` — Returns True only if **every** value is truthy

```python
all([1, True, "data"])   # True  — everything is truthy
all([1, 0, "data"])      # False — 0 is falsy, one failure fails all
```

> **Use case:** Validate that a user filled in ALL required form fields before saving.

---

### `any()` — Returns True if **at least one** value is truthy

```python
any([0, False, "data"])  # True  — one truthy value is enough
any([0, False, ""])      # False — nothing truthy at all
```

> **Use case:** Allow access if the user provided at least one contact method (email, phone, or username).

---

### `isinstance()` — Check whether a value belongs to a specific data type

```python
isinstance(42, int)       # True
isinstance(3.14, float)   # True
isinstance(True, str)     # False
```

> **Use case:** Validate incoming data before running calculations — e.g. confirm a price is a `float` before adding it to a total.

---

### String / Number methods that also return `True` / `False`

These work exactly like boolean expressions and can be used directly in conditions:

| Method | Example | What it checks |
|---|---|---|
| `.startswith()` | `"2026-07".startswith("2026")` | Does the string begin with this? |
| `.endswith()` | `"report.csv".endswith(".csv")` | Does the string end with this? |
| `.isnumeric()` | `"4820".isnumeric()` | Does the string contain only digits? |
| `.isalpha()` | `"Cairo".isalpha()` | Does the string contain only letters? |
| `.is_integer()` | `(7.0).is_integer()` | Does this float have no real decimals? |

---

## 2 — Comparison Operators

Comparison operators sit **between two values** and ask a yes/no question. The answer is always `True` or `False`.

| Operator | Meaning | Example | Result |
|---|---|---|---|
| `==` | Equal to | `10 == 10` | `True` |
| `!=` | Not equal to | `10 != 5` | `True` |
| `>` | Greater than | `7 > 3` | `True` |
| `>=` | Greater than or equal | `7 >= 7` | `True` |
| `<` | Less than | `3 < 7` | `True` |
| `<=` | Less than or equal | `3 <= 7` | `True` |

> ⚠️ **Common mistake:** `=` assigns a value. `==` compares two values. Never confuse them.

---

### Chain Comparison (Python's `BETWEEN`)

You can chain multiple comparisons in one line — Python evaluates left to right and **all parts must be True**:

```python
age = 20
18 <= age <= 30    # True  — age is between 18 and 30 (inclusive)
18 <= age <= 19    # False — age is outside this range
```

> **Use case:** Validate that a value falls within an acceptable range — e.g. a discount rate between 0% and 50%.

---

### Key rules

- Python compares **strings alphabetically**: `"a" < "b"` → `True`
- Python is **case-sensitive**: `"A" == "a"` → `False`
- You can compare **variables, values, expressions, or function results** — anything that returns a value

---

## 3 — Logical Operators

Logical operators **connect two or more boolean expressions** so you can evaluate them together.

---

### `and` — Both conditions must be True

```python
True and True    # True
True and False   # False
False and False  # False
```

```python
# Real use: login check — both email AND password must be correct
email_ok = True
password_ok = True
email_ok and password_ok    # True — access granted
```

---

### `or` — At least one condition must be True

```python
True or False    # True
False or False   # False — only case that returns False
```

```python
# Real use: alert system — CPU OR memory above threshold triggers alert
cpu_high    = False
memory_high = True
cpu_high or memory_high    # True — alert fires
```

---

### `not` — Flips True to False and False to True

```python
not True     # False
not False    # True
not 0        # True  (0 is falsy, not flips it)
not ""       # True  (empty string is falsy)
```

> **Use case:** Check that a value is NOT empty or NOT banned.

---

### Execution Order (Priority)

When mixing `and` and `or` in one line, Python executes `and` **before** `or` by default.

```python
# Without parentheses — and runs first, may give unexpected results
is_logged_in or is_guest and not is_banned

# With parentheses — you control the order
(is_logged_in or is_guest) and not is_banned   # correct logic
```

> **Rule:** Whenever you mix `and` with `or`, use **parentheses** to make the order explicit and avoid bugs.

---

## 4 — Membership Operators (`in` / `not in`)

Check whether a value exists **inside** a string, list, or any sequence.

| Operator | Returns True when… | Example |
|---|---|---|
| `in` | Value is found inside | `"@" in "user@gmail.com"` → `True` |
| `not in` | Value is NOT found | `"spam.com" not in safe_domains` → `True` |

```python
# Check if a transaction category is valid
valid_categories = ["Electronics", "Clothing", "Food"]
"Electronics" in valid_categories     # True
"Furniture"   in valid_categories     # False
```

> **Use case:** Validate that a user's email domain is not in a blacklist, or that a file extension is in an allowed list.

---

## 5 — Identity Operators (`is` / `is not`)

Check whether two variables point to the **same object in memory** — not whether their values are equal.

| Operator | Checks |
|---|---|
| `is` | Same object ID in memory |
| `is not` | Different object ID in memory |

```python
a = [1, 2, 3]
b = [1, 2, 3]

a == b     # True  — same values
a is b     # False — different objects in memory

c = a
a is c     # True  — c points to the same object as a
```

> **The one rule you must know:** Always use `is` and `is not` when comparing against `None`. Never use `==` with `None`.

```python
# Correct ✅
email is not None

# Avoid ❌
email != None
```

````python
# match-case is Python's answer to the switch statement (available from Python 3.10+).
# Syntax:
#   match variable:
#       case value1:  ...
#       case value2:  ...
#       case _:       ...  (wildcard — equivalent to "default" in other languages)
#
# Key advantage over if/elif: you can group multiple values in one case using |

# ── Example: Customer Support Ticket Router ──────────────────────────────────
# A data pipeline receives support tickets. Based on the issue category,
# match-case routes each ticket to the right team and sets the SLA in hours.

issue_category = input("Enter issue category (e.g. payment / delivery / account / product): ")

match issue_category.lower():          # .lower() in the match expression = case-insensitive
    case "payment" | "billing" | "invoice":
        team    = "Finance Team"
        sla_hrs = 4
    case "delivery" | "shipping" | "logistics":
        team    = "Operations Team"
        sla_hrs = 8
    case "product" | "quality" | "defect":
        team    = "Quality Assurance Team"
        sla_hrs = 24
    case "account" | "login" | "password":
        team    = "IT Support Team"
        sla_hrs = 2
    case _:                            # catches anything not listed above
        team    = "General Support"
        sla_hrs = 48

print(f"Ticket routed to : {team}")
print(f"SLA              : {sla_hrs} hours")
````

## 🔁 Iterable Objects in Python — Reference Guide for Data Analysis

An **iterable** is any object that can hand out its elements one at a time (it implements `__iter__`). Below is every category of iterable you will realistically loop over in a data analysis workflow, grouped by where it shows up in a real project.

### 1. Core Built-in Sequences
*The everyday containers you'll loop over constantly while cleaning and prepping data.*

| Object | Example | Real-life data analysis use case |
|---|---|---|
| `str` | `for ch in "Sales2024":` | Parsing raw text fields, validating product codes, cleaning messy string columns character by character |
| `list` | `for row in sales_list:` | Looping over a list of transactions, customer records, or scraped rows before loading them into a DataFrame |
| `tuple` | `for (name, price) in items:` | Iterating over fixed records like `(product, price)` pairs returned from a database query |
| `range` | `for i in range(len(df)):` | Generating index positions to loop over rows, build synthetic IDs, or repeat an operation N times |

### 2. Mapping & Set Types
*Used when your data is keyed (like a lookup table) or you care about uniqueness.*

| Object | Example | Real-life data analysis use case |
|---|---|---|
| `dict` | `for key, value in report.items():` | Looping over a JSON API response, a config of column-rename mappings, or aggregated `{category: total_sales}` results |
| `set` / `frozenset` | `for region in unique_regions:` | Iterating over deduplicated values — e.g. the distinct list of regions, customer IDs, or product categories in a dataset |

### 3. File & Raw Data I/O
*Every dataset starts life as a file — these iterators let you stream it in without loading everything to memory at once.*

| Object | Example | Real-life data analysis use case |
|---|---|---|
| File object (`open()`) | `for line in open("sales_log.txt"):` | Reading a large log file or raw text export line-by-line instead of loading the whole file into RAM |
| `csv.reader` / `csv.DictReader` | `for row in csv.DictReader(f):` | Reading a CSV export from an ERP/POS system row-by-row before building a DataFrame |
| Database cursor (`sqlite3`, `psycopg2`) | `for row in cursor.execute(query):` | Streaming query results from a SQL database (e.g. daily transactions table) without pulling millions of rows into memory at once |

### 4. Data Analysis Library Objects (pandas / NumPy)
*The iterables you'll use most once your data actually lives in analysis-ready structures.*

| Object | Example | Real-life data analysis use case |
|---|---|---|
| `pandas.Series` | `for value in df["revenue"]:` | Looping over a single column, e.g. flagging every revenue value below a threshold |
| `DataFrame.iterrows()` | `for idx, row in df.iterrows():` | Row-by-row business logic that's hard to vectorize, e.g. generating a custom invoice message per order |
| `DataFrame.itertuples()` | `for row in df.itertuples():` | Same as `iterrows()` but faster — preferred in production pipelines when looping row-by-row is unavoidable |
| `DataFrame.items()` | `for col_name, col_data in df.items():` | Looping over every column to apply a per-column data-quality check (nulls, dtype, outliers) |
| `numpy.ndarray` | `for x in np.array([10, 20, 30]):` | Iterating over numeric arrays for simulations, statistical sampling, or manual aggregation |

### 5. Functional & Lazy Iterators
*Wrap other iterables to combine, transform, or number them — extremely common in ETL scripts.*

| Object | Example | Real-life data analysis use case |
|---|---|---|
| `enumerate()` | `for i, row in enumerate(records):` | Numbering rows while looping — e.g. printing progress every 1,000 records processed |
| `zip()` | `for name, price in zip(names, prices):` | Combining two parallel lists (e.g. column of names + column of prices) pulled from separate sources |
| `map()` | `for cleaned in map(str.strip, raw_values):` | Applying a cleaning function (e.g. `str.strip`, `float`) across a whole column of raw values |
| `filter()` | `for r in filter(lambda x: x > 0, refunds):` | Looping only over rows that pass a business condition, e.g. only positive refund amounts |
| Generator / generator expression | `for chunk in read_in_chunks(file):` | Streaming huge datasets (multi-GB CSVs, API pagination) in manageable chunks without exhausting memory |

### Loop Control Keywords: `break`, `continue`, `pass`

*Sometimes you don't want a loop to simply run from start to finish — you want to cut it short, skip a step, or leave a branch deliberately empty. Python gives you three keywords for exactly that.*

| Keyword | What it does | Real-life data analysis use case |
|---|---|---|
| `break` | Immediately exits the **entire** loop — no more iterations run, even if items remain | Fail-fast validation: stop scanning a dataset the moment you hit a corrupted/invalid record |
| `continue` | Skips the **rest of the current iteration only** and jumps to the next one — the loop keeps going | Data cleaning: skip a missing/invalid value but keep processing every other row |
| `pass` | Does **nothing** — a placeholder so Python has a valid statement where one is syntactically required | Sketching out branch logic (e.g. an `if`/`elif`) before you've decided how to handle a particular case yet |

---

#### Demo 1 — `break`: stop scanning at the first corrupted record

```python
transactions = [120, 85, 43, -1, 67, 200]   # -1 = corrupted record sentinel

for amount in transactions:
    if amount < 0:
        print("Corrupted record found — stopping scan.")
        break
    print(f"Processing transaction: {amount} EGP")

print("Scan halted.")
```

**Expected output:**
```
Processing transaction: 120 EGP
Processing transaction: 85 EGP
Processing transaction: 43 EGP
Corrupted record found — stopping scan.
Scan halted.
```
Notice `67` and `200` are never processed — `break` exits the loop entirely, so the remaining transactions are never even looked at.

---

#### Demo 2 — `continue`: skip invalid values, keep the rest

```python
raw_ages = [25, None, 34, -5, 41, None, 29]

valid_ages = []
for age in raw_ages:
    if age is None or age < 0:
        continue
    valid_ages.append(age)

print("Valid ages collected:", valid_ages)
```

**Expected output:**
```
Valid ages collected: [25, 34, 41, 29]
```
Unlike `break`, the loop still visits **every** item in `raw_ages` — `continue` just skips the append step for the bad ones and moves straight to the next iteration.

#### Demo 3 — `pass`: a placeholder for logic that isn't ready yet

```python
order_status = ["shipped", "pending", "cancelled", "shipped", "pending"]

for status in order_status:
```

### While Loop — Condition-Based

*A `while` loop repeats its body for as long as a condition stays `True`. Unlike a `for` loop, it doesn't know in advance how many times it will run — Python re-checks the condition at the **top** of every iteration, and the loop stops naturally the moment it becomes `False`.*

**Real-life data analysis use case:** processing a queue of records until it's empty, retrying a data pull until you've made enough attempts, or accumulating transactions into a batch until you hit a size limit — anything where "keep going while X is still true" is a more natural fit than "repeat N times."

```python
pending_records = [101, 102, 103, 104, 105]   # record IDs waiting to be processed

while len(pending_records) > 0:
    current = pending_records.pop(0)
    print(f"Processing record ID {current}...")

print("All records processed. Queue is empty.")
```

**Expected output:**
```
Processing record ID 101...
Processing record ID 102...
Processing record ID 103...
Processing record ID 104...
Processing record ID 105...
All records processed. Queue is empty.
```
Every iteration, Python re-checks `len(pending_records) > 0` — once `pop(0)` has drained the list to empty, the condition becomes `False` and the loop ends on its own. No `break` was needed here.


````python
# ── Demo: condition-based while loop ─────────────────────────────────────────
pending_records = [101, 102, 103, 104, 105]

while len(pending_records) > 0:
    current = pending_records.pop(0)
    print(f"Processing record ID {current}...")

print("All records processed. Queue is empty.")
````

### While Loop — `while True`

*`while True:` has no built-in stopping condition — the condition is literally always `True`, so the loop is intentionally infinite. The only way out is an explicit `break` somewhere inside the body. This pattern is used exactly when you can't express "keep going while X" as a clean check at the top — instead, you check for the stop signal in the middle of the loop, as you encounter it.*

**Real-life data analysis use case:** reading a live/streaming feed (sensor readings, an API cursor, a log tail) where you don't know ahead of time how many records are coming, and you only find out it's time to stop when you actually hit a sentinel value or stop signal.

```python
sensor_readings = [22.4, 23.1, 21.8, None, 24.0]   # None = sensor disconnect signal
i = 0

while True:
    reading = sensor_readings[i]
    if reading is None:
        print("Sensor disconnected — stopping read loop.")
        break
    print(f"Reading: {reading}°C")
    i += 1

print("Reading loop ended.")
```

**Expected output:**
```
Reading: 22.4°C
Reading: 23.1°C
Reading: 21.8°C
Sensor disconnected — stopping read loop.
Reading loop ended.
```
Notice `24.0` is never read — the loop hits the `None` sentinel and `break`s out immediately, even though more readings technically exist after it in the list. Without that `break`, `while True` would run forever (or crash once `i` runs past the end of the list).


````python
# ── Demo: while True + break ─────────────────────────────────────────────────
sensor_readings = [22.4, 23.1, 21.8, None, 24.0]
i = 0

while True:
    reading = sensor_readings[i]
    if reading is None:
        print("Sensor disconnected — stopping read loop.")
        break
    print(f"Reading: {reading}°C")
    i += 1

print("Reading loop ended.")
````

---

#### Step 1 — Initialize the Batch
*Concepts: variables*

Before any loop, set up:
```python
settled_transactions = []

total_revenue = 0
category_totals = {}
best_txn_id, best_amount = None, 0
page_number = 0

```

---

#### Step 2 — Outer Loop: Page Fetcher
*Concepts: `while True`, `break` (Exit A)*

Write `while True:`. On every pass:
- **Exit A — out of pages:** if `page_number >= len(transaction_pages)`, print `"No more pages available."` and `break`.
- Otherwise, fetch `current_page = transaction_pages[page_number]`, print `f"Fetching page {page_number + 1}..."`, and add `1` to `pages_fetched`.

---

#### Step 3 — Inner Loop: Clean & Validate Each Transaction
*Concepts: nested `for`, string methods, type conversion, `continue`*

Loop `for txn in current_page:`. For each transaction:
- `customer_name` ← `txn["customer"].strip().title()`
- `category` ← `txn["category"].title()`
- `amount` ← `float(txn["amount"])`
- **Validation gate:** if `not customer_name or amount <= 0`, print `f"  Skipping {txn['txn_id']}: invalid record."`, add `1` to `skipped_count`, and `continue` to the next transaction.

---

#### Step 4 — Classify the Transaction
*Concepts: `if`, `match-case`, inline `if`, `pass`*

Still inside the inner loop, for every transaction that survives Step 3:
- If `category not in valid_categories`: use `pass` — there's no special handling for unrecognized categories yet, so it should just fall through and settle normally.
- Use `match-case` on `txn["payment"].title()` to set `processor`:

| Case | `processor` |
|---|---|
| `"Credit Card"` \| `"Debit Card"` | `"Stripe Gateway"` |
| `"Wallet"` \| `"Mobile Pay"` | `"PayMob Gateway"` |
| `"Cash"` | `"Manual Entry"` |
| `_` | `"Unknown"` |

- Set `priority` with an inline `if`: `"URGENT"` if `amount >= 2000` else `"STANDARD"`.

---

#### Step 5 — Settle & Accumulate
*Concepts: accumulator pattern, `dict` building, manual best-value comparison*

Still inside the inner loop:
- Append `txn["txn_id"]` to `settled_transactions`.
- Add `amount` onto `total_revenue`.
- Update `category_totals[category]` (add `amount`, creating the key with `.get(category, 0)` if it's new).
- If `amount > best_amount`, update `best_amount` and `best_txn_id` (manual comparison — no `max()`).
- Print `f"  Settled {txn['txn_id']} — {customer_name}, {amount} EGP, {priority}, via {processor}"`.
- **Exit B — quota reached:** if `len(settled_transactions) >= daily_cutoff`, set `cutoff_reached = True` and `break` (this breaks the inner `for` only).

---

#### Step 6 — Decide Whether to Fetch Another Page
*Concepts: `if`, flag-controlled `break`*

After the inner `for` loop ends, back in the outer `while`:
- If `cutoff_reached` is `True`: print `"Daily settlement cutoff reached."` and `break` (this is what actually stops the outer loop).
- Otherwise: add `1` to `page_number` and let `while True` loop again.
