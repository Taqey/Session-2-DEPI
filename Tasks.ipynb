```python
# ============================================
# Program Name : Dataset Profile Reporter
# Purpose      : Collect dataset information and generate a profile report.
# Author       : Your Name
# Date         : 21 July 2026
# ============================================

# ============================================
# Collect Dataset Information
# ============================================

# Stores the analyst's name
analyst_name = input("Enter your name: ")

# Stores the dataset name
dataset_name = input("Enter the dataset name: ")

# Stores the dataset source
data_source = input("Enter the data source (e.g. CSV, SQL, API): ")

# Stores the number of rows in the dataset
num_rows = int(input("Enter the number of rows: "))

# Stores the number of columns in the dataset
num_columns = int(input("Enter the number of columns: "))

# Stores the number of missing values
missing_values = int(input("Enter the number of missing values: "))

# ============================================
# Calculate Derived Metrics
# ============================================

# Calculates the total number of cells in the dataset
total_cells = num_rows * num_columns

# Calculates the percentage of missing values
missing_pct = (missing_values / total_cells) * 100

# ============================================
# Print Dataset Profile Report
# ============================================

print("=" * 44)
print(" " * 9 + "DATASET PROFILE REPORT")
print("=" * 44)

print(f"{'Analyst':<13}: {analyst_name}")
print(f"{'Dataset':<13}: {dataset_name}")
print(f"{'Source':<13}: {data_source}")

print("-" * 44)

print(f"{'Rows':<13}: {num_rows}")
print(f"{'Columns':<13}: {num_columns}")
print(f"{'Total Cells':<13}: {total_cells}")
print(f"{'Missing Vals':<13}: {missing_values}")
print(f"{'Missing %':<13}: {missing_pct:.2f}")

print("-" * 44)
print("   Profile generated successfully!")
print("=" * 44)

# ============================================
# Data Quality Verdict
# ============================================

if missing_pct < 5.0:
    print("Data Quality : GOOD  — missing data is within acceptable range.")
else:
    print("Data Quality : POOR  — missing data exceeds the acceptable threshold.")
```

```python
# ── Given Raw Data — do NOT change these ───────────────────────────────────
raw_name  = "   nour el-din   "           # extra spaces, wrong case
raw_price = "$1,299.50"                   # currency symbol and comma
raw_log   = "2026-07-01|TXN-8842|Electronics|Cairo|Completed"
raw_email = "  NOUR.ELDIN@Gmail.COM  "    # extra spaces, inconsistent case
raw_phone = "+20-(012)-345-6789"          # contains formatting characters

# ── Group 1 — Measurement ───────────────────────────────────────────────────
print("Raw Data Measurements:")
raw_name_length = len(raw_name)
print(f"Length of raw_name: {raw_name_length}")
 
name_length_after_strip = len(raw_name.strip())
print(f"Length of raw_name after stripping: {name_length_after_strip}")
print(f"whitespace removed: {raw_name_length - name_length_after_strip}")


print(f"Number of separators in raw_log: {raw_log.count('|')}")

# ── Group 2 — Cleaning ──────────────────────────────────────────────────────

clean_name = raw_name.strip().title()
print(f"Cleaned Name: {clean_name}")

clean_email = raw_email.strip().lower()
print(f"Cleaned Email: {clean_email}")

# ── Group 3 — Transformation ────────────────────────────────────────────────
unit_price=float(raw_price.replace("$", "").replace(",", ""))
print(f"Unit Price: {unit_price}")

clean_phone = (
    raw_phone
    .replace("+", "")
    .replace("(", "")
    .replace(")", "")
    .replace("-", "")
)

print(f"Cleaned Phone: {clean_phone}")

txn_date, txn_id, category, city, status = raw_log.split("|")
# print(f"Transaction Date: {txn_date}")
# print(f"Transaction ID: {txn_id}")
# print(f"Category: {category}")
# print(f"City: {city}")
# print(f"Status: {status}")

# ── Group 4 — Slicing & Indexing ────────────────────────────────────────────
txn_year = txn_date[:4]
txn_month = txn_date[5:7]
txn_day = txn_date[8:10]
print(f"Transaction Year: {txn_year}")
print(f"Transaction Month: {txn_month}")
print(f"Transaction Day: {txn_day}")

# ── Group 5 — Searching & Validation ────────────────────────────────────────
is_2026 = txn_date.startswith("2026")
has_at = "@" in clean_email
is_com = clean_email.endswith(".com")
phone_numeric = clean_phone.isnumeric()
first_separator = raw_log.find("|")

print(is_2026)
print(has_at)
print(is_com)
print(phone_numeric)
print(first_separator)

# ── Group 6 — f-string Summary Report ───────────────────────────────────────
print("=" * 48)
print("     ShopEasy Analytics — Cleaned Record")
print("=" * 48)

print("Name", clean_name, sep="\t\t\t: ")
print("Email", clean_email, sep="\t\t\t: ")
print("Phone", clean_phone, sep="\t\t\t: ")
print("Price", f"{unit_price} EGP", sep="\t\t\t: ")
print("Transaction", txn_id, sep="\t\t: ")
print("Category", category, sep="\t\t: ")
print("City", city, sep="\t\t\t: ")
print("Status", status, sep="\t\t\t: ")
print(
    "Date",
    f"{txn_date} (Year: {txn_year} | Month: {txn_month} | Day: {txn_day})",
    sep="\t\t\t: "
)

print("-" * 48)

print("Validation Checks:")

print("Date from 2026", is_2026, sep="\t\t: ")
print("Email has @", has_at, sep="\t\t: ")
print("Email ends .com", is_com, sep="\t\t: ")
print("Phone digits only", phone_numeric, sep="\t: ")

print("=" * 48)

```
