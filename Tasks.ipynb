```python
store_name   = "FreshMart"
cashier_name = "Sara"

item1_name  = "Apples"
item1_price = 12.50

item2_name  = "Bread"
item2_price = 8.75

item3_name  = "Milk"
item3_price = 15.00

total = item1_price + item2_price + item3_price

print("=" * 32)
print(store_name.center(32))
print("=" * 32)
print("Item", "Price", sep="\t\t\t")
print("-" * 32)

print(item1_name, f"{item1_price:.2f} EGP", sep="\t\t\t")
print(item2_name, f"{item2_price:.2f} EGP", sep="\t\t\t")
print(item3_name, f"{item3_price:.2f} EGP", sep="\t\t\t")

print("-" * 32)
print("TOTAL", f"{total:.2f} EGP", sep="\t\t\t")
print("=" * 32)
print("Cashier:", cashier_name)
print("Thank you for shopping with us!")
print("=" * 32)
```

```python
company_name = "ShopEasy Analytics"
report_month = "June 2026"
total_orders = 1482
total_revenue = 287450.75
product_return_rate = 3.8
top_selling_category = "Electronics"
monthly_target_achieved = True
avg_order_value = total_revenue / total_orders

print("="*42)
print("   " f"{company_name}")
print("   " f"Monthly Performance Report | {report_month}")
print("="*42)
print("metric", "value", sep="\t\t\t\t")
print("-"*42)
print("Total Orders", f"{total_orders}", sep="\t\t\t")
print("Total Revenue", f"{total_revenue:.2f} EGP", sep="\t\t\t")
print("Average Order Value", f"{avg_order_value:.2f} EGP", sep="\t\t")
print("Product Return Rate", f"{product_return_rate:.2f}%", sep="\t\t")
print("Top Selling Category", f"{top_selling_category}", sep="\t\t")
print("Monthly Target Achieved", f"{monthly_target_achieved}", sep="\t\t")
print("="*42)
print(
(    "Great job - Target achieved!"
      if monthly_target_achieved 
      else "Keep pushing - Target not met.".center(42)).center(42)
    )
print("="*42)
```
