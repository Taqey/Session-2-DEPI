```python
# ── Given Order Record — do NOT change these ────────────────────────────────
order_id         = "ORD-20260612-007"
customer_tier    = "Gold"
order_amount     = 2750.00
num_items        = 8
payment_method   = "Credit Card"
delivery_region  = "Alexandria"
is_first_order   = False
loyalty_points   = 850
coupon_code      = "SAVE15"
days_to_deliver  = 3
stock_available  = True

# ── Step 1 — Discount Rate Classifier (if / elif / else + ==) ───────────────

if customer_tier == "Platinum":
    discount_rate = 20
elif customer_tier == "Gold":
    discount_rate = 15
elif customer_tier == "Silver":
    discount_rate = 10
elif customer_tier == "Bronze":
    discount_rate = 5
else:
    discount_rate = 0
    print("⚠ Unknown tier — no discount applied")

discount_amount = order_amount * discount_rate / 100

# ── Step 2 — First Order Bonus (nested if / else) ───────────────────────────

if is_first_order:
    if order_amount > 1000:
        discount_rate += 5
        print("First order bonus: +5% applied")
    else:
        discount_rate += 3
        print("First order bonus: +3% applied")
else:
    pass

discount_amount = order_amount * discount_rate / 100

# ── Step 3 — Coupon Validation (is not None + bool() + if / elif / else) ────

if coupon_code is not None and bool(coupon_code):
    if coupon_code == "SAVE15":
        coupon_discount = 15
    elif coupon_code == "SAVE30":
        coupon_discount = 30
    elif coupon_code == "SAVE50":
        coupon_discount = 50
    else:
        coupon_discount = 0
        print("⚠ Invalid coupon code")
else:
    coupon_discount = 0

# ── Step 4 — Shipping Cost Classifier (if / elif / else + or + nested if) ───

if delivery_region == "Abroad":
    shipping_cost = 150
elif delivery_region == "Aswan":
    shipping_cost = 60
elif delivery_region == "Cairo" or delivery_region == "Alexandria":
    if order_amount >= 500:
        shipping_cost = 0
    else:
        shipping_cost = 30
else:
    shipping_cost = 30

# ── Step 5 — Priority Flag (and / or / in / != on condition lines) ──────────

if (
    order_amount > 2000
    and num_items > 5
    and customer_tier in ["Gold", "Platinum"]
):
    priority_flag = "URGENT"
elif order_amount > 1000 or loyalty_points > 500:
    priority_flag = "HIGH"
elif stock_available and payment_method != "Cash":
    priority_flag = "STANDARD"
else:
    priority_flag = "HOLD"

# ── Step 6 — Final Approval Decision (isinstance + and + separate if checks) ─

if (
    isinstance(order_amount, (int, float))
    and order_amount > 0
    and stock_available
    and days_to_deliver <= 7
):
    decision = "APPROVED"
else:
    if not isinstance(order_amount, (int, float)):
        print("⚠ Rejection reason: amount is not numeric")

    if isinstance(order_amount, (int, float)) and order_amount <= 0:
        print("⚠ Rejection reason: amount is zero or negative")

    if not stock_available:
        print("⚠ Rejection reason: out of stock")

    if days_to_deliver > 7:
        print("⚠ Rejection reason: delivery exceeds 7 days")

    decision = "REJECTED"

# ── Step 7 — Calculate Final Amount & Print Report (f-strings) ──────────────

final_amount = order_amount - discount_amount - coupon_discount + shipping_cost

print("=" * 52)
print("    ShopEasy Analytics — Order Processing Report")
print("=" * 52)

print(f"Order ID         : {order_id}")
print(f"Customer Tier    : {customer_tier}")
print(f"Payment          : {payment_method}")
print(f"Delivery Region  : {delivery_region}")

print("-" * 52)

print("Pricing Breakdown")
print(f"  Original Amount  : {order_amount} EGP")
print(f"  Discount Rate    : {discount_rate}%")
print(f"  Discount Amount  : {discount_amount} EGP")
print(f"  Coupon Saving    : {coupon_discount} EGP")
print(f"  Shipping Cost    : {shipping_cost} EGP")
print(f"  Final Amount     : {final_amount} EGP")

print("-" * 52)

print("Order Details")
print(f"  Items            : {num_items}")
print(f"  Days to Deliver  : {days_to_deliver}")
print(f"  First Order      : {is_first_order}")
print(f"  Stock Available  : {stock_available}")

print("-" * 52)

print("Classification")
print(f"  Priority Flag    : {priority_flag}")
print(f"  Decision         : {decision}")

print("=" * 52)
```
