```python
# ── Given Order Record — do NOT change these ────────────────────────────────
order_id         = "ORD-20260612-007"
customer_tier    = "Gold"
order_amount     = 2750.00
num_items        = 8
payment_method   = "Credit Card"
delivery_region  = "Alexandria"
is_first_order   = False
loyalty_points   = 850
coupon_code      = "SAVE15"
days_to_deliver  = 3
stock_available  = True

# ── Step 1 — Discount Rate Classifier (if / elif / else + ==) ───────────────

if customer_tier == "Platinum":
    discount_rate = 20
elif customer_tier == "Gold":
    discount_rate = 15
elif customer_tier == "Silver":
    discount_rate = 10
elif customer_tier == "Bronze":
    discount_rate = 5
else:
    discount_rate = 0
    print("⚠ Unknown tier — no discount applied")

discount_amount = order_amount * discount_rate / 100

# ── Step 2 — First Order Bonus (nested if / else) ───────────────────────────

if is_first_order:
    if order_amount > 1000:
        discount_rate += 5
        print("First order bonus: +5% applied")
    else:
        discount_rate += 3
        print("First order bonus: +3% applied")
else:
    pass

discount_amount = order_amount * discount_rate / 100

# ── Step 3 — Coupon Validation (is not None + bool() + if / elif / else) ────

if coupon_code is not None and bool(coupon_code):
    if coupon_code == "SAVE15":
        coupon_discount = 15
    elif coupon_code == "SAVE30":
        coupon_discount = 30
    elif coupon_code == "SAVE50":
        coupon_discount = 50
    else:
        coupon_discount = 0
        print("⚠ Invalid coupon code")
else:
    coupon_discount = 0

# ── Step 4 — Shipping Cost Classifier (if / elif / else + or + nested if) ───

if delivery_region == "Abroad":
    shipping_cost = 150
elif delivery_region == "Aswan":
    shipping_cost = 60
elif delivery_region == "Cairo" or delivery_region == "Alexandria":
    if order_amount >= 500:
        shipping_cost = 0
    else:
        shipping_cost = 30
else:
    shipping_cost = 30

# ── Step 5 — Priority Flag (and / or / in / != on condition lines) ──────────

if (
    order_amount > 2000
    and num_items > 5
    and customer_tier in ["Gold", "Platinum"]
):
    priority_flag = "URGENT"
elif order_amount > 1000 or loyalty_points > 500:
    priority_flag = "HIGH"
elif stock_available and payment_method != "Cash":
    priority_flag = "STANDARD"
else:
    priority_flag = "HOLD"

# ── Step 6 — Final Approval Decision (isinstance + and + separate if checks) ─

if (
    isinstance(order_amount, (int, float))
    and order_amount > 0
    and stock_available
    and days_to_deliver <= 7
):
    decision = "APPROVED"
else:
    if not isinstance(order_amount, (int, float)):
        print("⚠ Rejection reason: amount is not numeric")

    if isinstance(order_amount, (int, float)) and order_amount <= 0:
        print("⚠ Rejection reason: amount is zero or negative")

    if not stock_available:
        print("⚠ Rejection reason: out of stock")

    if days_to_deliver > 7:
        print("⚠ Rejection reason: delivery exceeds 7 days")

    decision = "REJECTED"

# ── Step 7 — Calculate Final Amount & Print Report (f-strings) ──────────────

final_amount = order_amount - discount_amount - coupon_discount + shipping_cost

print("=" * 52)
print("    ShopEasy Analytics — Order Processing Report")
print("=" * 52)

print(f"Order ID         : {order_id}")
print(f"Customer Tier    : {customer_tier}")
print(f"Payment          : {payment_method}")
print(f"Delivery Region  : {delivery_region}")

print("-" * 52)

print("Pricing Breakdown")
print(f"  Original Amount  : {order_amount} EGP")
print(f"  Discount Rate    : {discount_rate}%")
print(f"  Discount Amount  : {discount_amount} EGP")
print(f"  Coupon Saving    : {coupon_discount} EGP")
print(f"  Shipping Cost    : {shipping_cost} EGP")
print(f"  Final Amount     : {final_amount} EGP")

print("-" * 52)

print("Order Details")
print(f"  Items            : {num_items}")
print(f"  Days to Deliver  : {days_to_deliver}")
print(f"  First Order      : {is_first_order}")
print(f"  Stock Available  : {stock_available}")

print("-" * 52)

print("Classification")
print(f"  Priority Flag    : {priority_flag}")
print(f"  Decision         : {decision}")

print("=" * 52)
```

```python
import math
import random

# ── Given Raw Export — do NOT change these ──────────────────────────────────
raw_txn_id   = "  TXN-20260701-9934  "
raw_customer = "  nour el-din  "
raw_email    = "  NOUR.ELDIN@Gmail.COM  "
raw_phone    = "+20-(010)-987-6543"
raw_amount   = "$3,199.00"
raw_items    = "5"
raw_discount = "20"
raw_category = "ELECTRONICS"
raw_region   = "  Cairo  "
raw_payment  = "credit card"
raw_date     = "2026-07-01"
raw_tier     = "  Platinum  "
raw_verified = "True"
raw_source   = "None"

valid_categories = ["Electronics", "Clothing", "Food", "Books", "Beauty"]
banned_domains   = ["spam.com", "fake.org", "temp.net", "trash.io"]

# ══════════════════════════════════════════════════════════════════════════════
# PHASE 1 — Analyst Session Setup (comments + variables + input + random)
# ══════════════════════════════════════════════════════════════════════════════

# Program Name : Transaction Data Cleaner
# Purpose      : Clean and validate transaction records for analysis
# Author       : Taqeyy Eldeen Hussien
# Date         : 23 July 2026

analyst_name = input("Enter your name: ")
shift = input("Enter your shift (morning/afternoon/evening): ")
session_id = random.randint(10000, 99999)

# ══════════════════════════════════════════════════════════════════════════════
# PHASE 2 — Data Cleaning (string methods — all 6 groups)
# ══════════════════════════════════════════════════════════════════════════════

# ── Group 1 — Measurement ────────────────────────────────────────────────────
len_before_strip = len(raw_customer)
len_after_strip = len(raw_customer.strip())
print(f"Hidden spaces: {len_before_strip - len_after_strip}")

date_separators = raw_date.count("-")
print(f"Date separators found: {date_separators}")

# ── Group 2 — Cleaning (whitespace + case) ───────────────────────────────────
txn_id = raw_txn_id.strip()
customer_name = raw_customer.strip().title()
customer_email = raw_email.strip().lower()
region = raw_region.strip().title()
customer_tier = raw_tier.strip().title()

# ── Group 3 — Transformation (restructure) ───────────────────────────────────
order_amount = float(raw_amount.replace("$", "").replace(",", ""))
clean_phone = raw_phone.replace("+", "").replace("(", "").replace(")", "").replace("-", "")
payment_method = raw_payment.title()
product_category = raw_category.title()

# ── Group 4 — Slicing (extract date parts) ───────────────────────────────────
txn_year = raw_date[:4]
txn_month = raw_date[5:7]
txn_day = raw_date[-2:]

# ── Group 5 — Searching (structural checks) ──────────────────────────────────
date_from_2026 = raw_date.startswith("2026")
email_has_at = "@" in customer_email
email_ends_com = customer_email.endswith(".com")
phone_digits = clean_phone.isnumeric()

print(f"Date starts with 2026: {date_from_2026}")
print(f"Email has @: {email_has_at}")
print(f"Email ends .com: {email_ends_com}")
print(f"Phone numeric: {phone_digits}")

# ── Group 6 — Validation (content checks) ────────────────────────────────────
items_numeric = raw_items.isnumeric()
print(f"Items numeric: {items_numeric}")

# ══════════════════════════════════════════════════════════════════════════════
# PHASE 3 — Type Conversion (data types + type() + isinstance)
# ══════════════════════════════════════════════════════════════════════════════

num_items = int(raw_items)
print(type(num_items))

discount_pct = int(raw_discount)
print(type(discount_pct))

is_verified = raw_verified == "True"
print(type(is_verified))

data_source = None if raw_source == "None" else raw_source
print(type(data_source))

# ══════════════════════════════════════════════════════════════════════════════
# PHASE 4 — Quality Gates (bool + all + any + isinstance + comparison + logical
#                          + membership + identity)
# ══════════════════════════════════════════════════════════════════════════════

name_ok = bool(customer_name)
amount_ok = isinstance(order_amount, float)
range_ok = order_amount > 0 and 0 <= discount_pct <= 50 and num_items > 0
category_ok = product_category in valid_categories

email_domain = customer_email.split("@")[1]
domain_safe = email_domain not in banned_domains

email_ok = bool(customer_email) and "@" in customer_email and customer_email.endswith(".com")
source_flagged = data_source is None
all_fields_ok = all([name_ok, amount_ok, category_ok, email_ok])
any_warnings = any([not range_ok, source_flagged, not domain_safe])

# ══════════════════════════════════════════════════════════════════════════════
# PHASE 5 — Classification Engine (if/elif/else + match-case + inline if)
# ══════════════════════════════════════════════════════════════════════════════

# ── 5a — Amount Tier (if / elif / else) ──────────────────────────────────────
if order_amount >= 3000:
    amount_tier = "Whale"
elif order_amount >= 1000:
    amount_tier = "High Value"
elif order_amount >= 500:
    amount_tier = "Mid Value"
else:
    amount_tier = "Low Value"

# ── 5b — Payment Processor (match-case) ──────────────────────────────────────
match payment_method:
    case "Credit Card" | "Debit Card":
        processor = "Stripe Gateway"
        fee_pct = 2.5
    case "Wallet" | "Mobile Pay":
        processor = "PayMob Gateway"
        fee_pct = 1.0
    case "Cash":
        processor = "Manual Entry"
        fee_pct = 0.0
    case _:
        processor = "Unknown"
        fee_pct = 3.0

# ── 5c — Priority & Source Label (inline if) ─────────────────────────────────
priority = "URGENT" if order_amount >= 2000 and customer_tier in ["Gold", "Platinum"] else "STANDARD"
source_label = "⚠ Unknown Source" if data_source is None else "✓ Verified"

# ══════════════════════════════════════════════════════════════════════════════
# PHASE 6 — Financial Calculations (numeric operators + math module)
# ══════════════════════════════════════════════════════════════════════════════

discount_amount = order_amount * discount_pct / 100
processing_fee = round(order_amount * fee_pct / 100, 2)
net_amount = order_amount - discount_amount
final_amount = net_amount + processing_fee
tax_amount = round(final_amount * 0.14, 2)
total_payable = round(final_amount + tax_amount, 2)
savings = round(abs(order_amount - total_payable), 2)
floor_estimate = math.floor(final_amount)
ceil_estimate = math.ceil(final_amount)

# ══════════════════════════════════════════════════════════════════════════════
# PHASE 7 — Full Audit Report (f-strings + print)
# ══════════════════════════════════════════════════════════════════════════════

print("=" * 64)
print("         ShopEasy Analytics — Transaction Audit Report")
print(f"                     Session: {session_id}")
print("=" * 64)
print(f"Analyst          : {analyst_name}")
print(f"Shift            : {shift}")
print("=" * 64)
print("TRANSACTION RECORD")
print(f"  ID               : {txn_id}")
print(f"  Date             : {raw_date}  (Y: {txn_year} | M: {txn_month} | D: {txn_day})")
print(f"  Customer         : {customer_name}")
print(f"  Email            : {customer_email}")
print(f"  Phone            : {clean_phone}")
print(f"  Region           : {region}")
print(f"  Tier             : {customer_tier}")
print("-" * 64)
print("PRODUCT & PAYMENT")
print(f"  Category         : {product_category}")
print(f"  Items            : {num_items}")
print(f"  Payment Method   : {payment_method}")
print(f"  Processor        : {processor}")
print(f"  Processing Fee   : {fee_pct}%")
print("-" * 64)
print("QUALITY GATES")
print(f"  G1 Name exists   : {name_ok}")
print(f"  G2 Amount float  : {amount_ok}")
print(f"  G3 Range valid   : {range_ok}")
print(f"  G4 Category ok   : {category_ok}")
print(f"  G5 Domain safe   : {domain_safe}")
print(f"  G6 Email valid   : {email_ok}")
print(f"  G7 Source flagged: {source_flagged}")
print(f"  G8 All required  : {all_fields_ok}")
print(f"  G9 Any warnings  : {any_warnings}")
print("-" * 64)
print("CLASSIFICATION")
print(f"  Amount Tier      : {amount_tier}")
print(f"  Priority         : {priority}")
print(f"  Source Status    : {source_label}")
print("-" * 64)
print("FINANCIALS (EGP)")
print(f"  Original Amount  : {order_amount}")
print(f"  Discount ({discount_pct}%)   : -{discount_amount}")
print(f"  Processing Fee   : +{processing_fee}")
print(f"  Net Amount       : {net_amount}")
print(f"  VAT (14%)        : +{tax_amount}")
print(f"  Total Payable    : {total_payable}")
print(f"  Customer Savings : {savings}")
print(f"  Floor Estimate   : {floor_estimate}")
print(f"  Ceil Estimate    : {ceil_estimate}")
print("-" * 64)
print("VALIDATION SUMMARY")
print(f"  Date from 2026   : {date_from_2026}")
print(f"  Email has @      : {email_has_at}")
print(f"  Email ends .com  : {email_ends_com}")
print(f"  Phone digits     : {phone_digits}")
print(f"  Items numeric    : {items_numeric}")
print("=" * 64)

if any_warnings:
    print("⚠ Pipeline complete — record flagged for manual review before warehouse commit.")
else:
    print("✓ Pipeline complete — record cleared for automatic warehouse commit.")
```

```python
# ── Given Data — do NOT change these ────────────────────────────────────────
day_names    = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
daily_sales  = [15200, 18400, 9800, 21300, 17650, 24900, 12100]
daily_target = 15000

# ── Steps 1-4 — Loop & Print / Running Total / Best & Worst / Days Above Target ──
weekly_total = 0
days_above_target = 0

best_day = day_names[0]
best_sales = daily_sales[0]
worst_day = day_names[0]
worst_sales = daily_sales[0]

print("=" * 50)
print("       ShopEasy Analytics — Weekly Sales Trend")
print("=" * 50)

for index, sales in enumerate(daily_sales):
    day = day_names[index]

    # Step 1 — Print each day and sales
    print(f"{day:<12} : {sales:6} EGP")

    # Step 2 — Running total
    weekly_total += sales

    # Step 3 — Best & Worst day
    if sales > best_sales:
        best_sales = sales
        best_day = day
    elif sales < worst_sales:
        worst_sales = sales
        worst_day = day

    # Step 4 — Days above target
    if sales > daily_target:
        days_above_target += 1

# ── Step 5 — Mini Bar Chart ─────────────────────────────────────────────────
print("-" * 50)
print("MINI BAR CHART")
print("-" * 50)

for day, sales in zip(day_names, daily_sales):
    bar = "*" * (sales // 2000)
    print(f"{day:<10}: {bar} ({sales})")

# ── Step 6 — Weekly Average & Final Report ──────────────────────────────────
weekly_average = round(weekly_total / len(daily_sales), 2)

print("-" * 50)
print(f"Weekly Total     : {weekly_total:6} EGP")
print(f"Weekly Average   : {weekly_average} EGP")
print(f"Best Day         : {best_day} ({best_sales} EGP)")
print(f"Worst Day        : {worst_day} ({worst_sales} EGP)")
print(f"Days Above Target: {days_above_target} / {len(daily_sales)}")
print("=" * 50)

# ── Bonus Challenge — Critical days below half target ───────────────────────
critical_days = 0

for sales in daily_sales:
    if sales < (daily_target / 2):
        critical_days += 1

if critical_days > 0:
    print(f"⚠ Warning: {critical_days} critical day(s) had sales below half the target.")

# ── Extension — Practice break ──────────────────────────────────────────────
critical_threshold = daily_target - 5000   # 10,000 EGP
critical_day = None

for day, sales in zip(day_names, daily_sales):
    print(f"Scanning {day}...")

    if sales < critical_threshold:
        critical_day = day
        print(f"🚨 Critical day detected: {critical_day} — halting scan.")
        break

if critical_day is None:
    print("✅ No critical days this week.")
```

```python
# ── Given Data — do NOT change these ────────────────────────────────────────
feedback_list = [
    "great service and fast delivery",
    "the app was slow and the checkout was slow",
    "great prices and great support",
    "delivery was late and packaging was poor",
    "fast delivery and great app experience",
]
negative_keywords = ["slow", "late", "poor"]

# ── Step 1 — Word Counter (nested for + dict accumulation) ──────────────────
word_counts = {}
total_words = 0

for review in feedback_list:
    for word in review.split():
        total_words += 1

        if word in word_counts:
            word_counts[word] += 1
        else:
            word_counts[word] = 1

# ── Step 2 — Most Frequent Word (manual comparison over dict.items()) ───────
most_common_word = ""
most_common_count = 0

for word, count in word_counts.items():
    if count > most_common_count:
        most_common_word = word
        most_common_count = count

# ── Step 3 — Negative Keyword Scanner (nested for + in + break) ─────────────
flagged_reviews = []
negative_review_count = 0

for i, review in enumerate(feedback_list):
    for keyword in negative_keywords:
        if keyword in review:
            flagged_reviews.append(
                f"Review {i + 1}: contains '{keyword}'"
            )
            negative_review_count += 1
            break

# ── Bonus — Shortest Review (manual comparison, no min()) ───────────────────
shortest_review = feedback_list[0]
shortest_length = len(shortest_review.split())

for review in feedback_list:
    review_length = len(review.split())

    if review_length < shortest_length:
        shortest_review = review
        shortest_length = review_length

# ── Step 4 — Averages & Final Report ────────────────────────────────────────
average_words_per_review = round(
    total_words / len(feedback_list), 2
)

print("=" * 50)
print("    ShopEasy Analytics — Feedback Scan Report")
print("=" * 50)
print(f"Total Reviews Scanned   :  {len(feedback_list)}")
print(f"Total Words             :  {total_words}")
print(f"Avg Words / Review      :  {average_words_per_review}")
print("-" * 50)
print(
    f"Most Mentioned Word     :  '{most_common_word}'  "
    f"({most_common_count} times)"
)
print("-" * 50)
print(f"Flagged Reviews         :  {negative_review_count}")

for flagged in flagged_reviews:
    print(" ", flagged)

print("-" * 50)
print("Shortest Review         :")
print(f"  \"{shortest_review}\" ({shortest_length} words)")
print("=" * 50)

# ── Extension — Practice continue ───────────────────────────────────────────
long_reviews = []

for review in feedback_list:
    words = review.split()

    if len(words) < 6:
        continue

    long_reviews.append(review)

print()
print("Long reviews kept:", len(long_reviews), "/", len(feedback_list))

for r in long_reviews:
    print(" -", r)
```

```python
# ── Given Data — do NOT change these ────────────────────────────────────────
api_pages = [
    [101, 102, 103],
    [104, 105, 106],
    [107, 108],
    [109, 110, 111],
    [112, 113, 114],
]
max_records = 10

# ── Step 1 — Paginate with while True (two break exit conditions) ───────────

collected_records = []
page_number = 0
pages_fetched = 0

while True:

    # Exit condition A — no more pages
    if page_number >= len(api_pages):
        print("No more pages available.")
        break

    current_page = api_pages[page_number]
    print(f"Fetching page {page_number + 1}...")

    for record_id in current_page:
        collected_records.append(record_id)

    pages_fetched += 1

    # Exit condition B — quota reached
    if len(collected_records) >= max_records:
        print("Record quota reached.")
        break

    page_number += 1


# ── Step 2 — Final Report ───────────────────────────────────────────────────

print("-" * 50)
print(f"Total records collected:  {len(collected_records)}")
print(f"Pages fetched          :  {pages_fetched}")
print(f"Record IDs             :  {collected_records}")
```
